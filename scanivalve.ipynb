```python
import numpy as np
import matplotlib.pyplot as plt
import scanivalve
```

```python
s = scanivalve.Scanivalve()
```

```python
s.config(FPS=10000, AVG=1, PERIOD=200)
```

```python
s.start()
```

```python
s.stop()
```

```python
p,f = s.read()

plt.plot(1/f * np.arange(p.shape[0]), p[:,0])
```
